```python
incidentes = pd.read_hdf( 'c:/_DATA/DW3/INCIDENTES.h5' )
incidentes.head( 3 )
```

```python
def make_point( row ) :
    return Point( (row['LONGITUD'], row['LATITUD']) ) 


def add_basemap(ax, zoom, url='http://tile.stamen.com/terrain/tileZ/tileX/tileY.png'):
    xmin, xmax, ymin, ymax = ax.axis()
    basemap, extent = ctx.bounds2img(xmin, ymin, xmax, ymax, zoom=zoom, url=url)
    ax.imshow(basemap, extent=extent, interpolation='bilinear')
    # restore original x/y limits
    ax.axis((xmin, xmax, ymin, ymax))
```

```python
incidentes['Coordinates'] = incidentes.apply( make_point, axis=1 )
incidentes.head( 3 )

geo_incidentes = geopandas.GeoDataFrame( incidentes, geometry='Coordinates' ) 
# type( geo_incidentes ) = geopandas.geodataframe.GeoDataFrame
geo_incidentes.crs = {'init' :'epsg:4326'}  
geo_incidentes = geo_incidentes.to_crs( epsg=3857 )

geo_incidentes.head() 
```

```python
help( ctx.add_basemap )
```

```python
barrios_med = barrios_med.to_crs( epsg=3857 )

ax_med = barrios_med.plot(figsize=(10, 10), alpha=0.3, edgecolor='k')
ax_med.set_xlim((-8419000, -8408000))
ax_med.set_ylim((  690000,   704000))
ax_med.set_xticklabels( [] ) 
ax_med.set_yticklabels( [] ) 
geo_incidentes.plot( ax = ax_med, color='red', markersize=2 )

#add_basemap(ax_med, zoom=13,  url="https://b.tile.openstreetmap.org/tileZ/tileX/tileY.png" ) 
ctx.add_basemap( ax_med, url="https://a.tile.openstreetmap.org/tileZ/tileX/tileY.png"  )
# url='http://toolserver.org/~cmarqu/hill/tileZ/tileX/tileY.png' )
plt.show() 
```
